```python
import zipfile
import os

def zip_directory(folder_path, zip_name):
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, start=folder_path)
                zipf.write(file_path, arcname)

# Usage
zip_directory('/content/runs', '10 freeze runs.zip')
```

```python
import os
import shutil
```

```python
created = os.listdir("/content/esrgan_images")
```

```python
import cv2
from tqdm import tqdm
```

```python
input_folder = f"dataset/images/test"

image_files = [f for f in os.listdir(input_folder) if f.endswith(".jpg") or f.endswith(".png")]

for filename in tqdm(image_files):
    path = os.path.join(input_folder, filename)
    img = cv2.imread(path)
    if img is None:
        print(f"Hata: {filename} okunamadı.")
        continue

    # 1. Tüm görselleri 640x640'e normalize et
    base_img = cv2.resize(img, (640, 640), interpolation=cv2.INTER_AREA)
    cv2.imwrite(os.path.join("/content/dataset/images/640",filename), base_img)
```

```python
src_folder = "/content/dataset/images/test"
dst_folder = "/content/dataset/images/sr"

# Klasör yoksa oluştur
os.makedirs(dst_folder, exist_ok=True)

# Dosya listesini sırala ve ilk 500 tanesini al
all_files = sorted(os.listdir(src_folder))

# Dosyaları kopyala
i = 0
j = 0
while j<len(all_files):
    if all_files[j].replace(".jpg","_out.jpg") not in created:
        src_path = os.path.join(src_folder, all_files[j])
        dst_path = os.path.join(dst_folder, all_files[j])
        shutil.copy(src_path, dst_path)
        i+=1
    j+=1
print(len(os.listdir(dst_folder)))
```
