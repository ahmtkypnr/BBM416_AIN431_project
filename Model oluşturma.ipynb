```python
import json
import gc
from torch import cuda
from ultralytics import YOLO
from BBM416_AIN431_project.data import create_shuffle_dataset, exclude_classes
```

```python
create_shuffle_dataset() # iÃ§inde Ã§alÄ±ÅŸtÄ±ÄŸÄ±nÄ±z dizine kurulum yapar, Ã¶zel dizin belirtebilirsiniz (dataset_path = ...)
```

```python
with open('/content/dataset/classes.json') as f:
  classes = json.load(f)
```

```python
classes_to_exclude = []
for class_, item in sorted(classes.items(), key=lambda x: len(x[1]), reverse=True):
    print(class_, len(item))
    if len(item) < 100:
      classes_to_exclude.append(class_)
exclude_classes(classes_to_exclude, "/content/dataset")
```

```python
model = YOLO('yolo11s.pt')
```

```python
model.train(
    data='/content/dataset/data_excluded.yaml',
    device=0,
    epochs=40,
    imgsz=640,
    batch=64,
    freeze=0
)
```

```python
def train_model(name='yolo11s.pt', epoch=40, size=640, batch=64, freeze=0, lr=None):
    model = YOLO(name)

    if lr:
        optimizer = 'AdamW'
    else:
        optimizer = 'auto'

    model.train(
        data='/content/dataset/data_excluded.yaml',
        device=0,
        epochs=epoch,
        imgsz=size,
        batch=batch,
        freeze=freeze,
        optimizer=optimizer,
        lr0=lr
    )

    del model
    gc.collect()
    cuda.empty_cache()
```

```python
train_model(freeze=5)
```

```python
train_model(freeze=10)
```

```python
train_model(batch=32, freeze=5)
```

```python
import zipfile
import os

def zip_directory(folder_path, zip_name):
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, start=folder_path)
                zipf.write(file_path, arcname)

# Usage
zip_directory('/content/runs', '10 freeze runs.zip')
```
